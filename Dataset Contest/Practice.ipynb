```python
#train data work
```

```python
train = train.drop(['Cabin', 'PassengerId', 'Name', 'Ticket'], axis =1)
```

```python
#handle null value
```

```python
train.isnull().sum()
```

```python
train['Age'] = train['Age'].fillna(train['Age'].mean())
```

```python
train['Embarked'] = train['Embarked'].fillna(train['Embarked'].mode()[0])
```

```python
#Encode numerical value
```

```python
train.info()
```

```python
train = pd.get_dummies(train, columns = ['Sex', 'Embarked'], dtype = int)
```

```python
X = train.drop('Survived', axis = 1)
```

```python
Y = train['Survived']
```

```python
X.head()
```

```python
Y.head()
```

```python
#test data work
```

```python
test
```

```python
test = test.drop(['Cabin', 'PassengerId', 'Name', 'Ticket'], axis =1)
```

```python
test.isnull().sum()
```

```python
test['Age'] = test['Age'].fillna(test['Age'].mean())
```

```python
test['Fare'] = test['Fare'].fillna(test['Fare'].mean())
```

```python
test.info()
```

```python
test = pd.get_dummies(test, columns = ['Sex', 'Embarked'], dtype = int)
```

```python
test
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression()
reg.fit(X_train, y_train)
y_pred = reg.predict(X_test)
accuracy_reg = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy_reg *100:.2f}%")
```

```python
from sklearn.ensemble  import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
y_pred2 = rf.predict(X_test)
accuracy_rf = accuracy_score(y_test,y_pred2)
print(f"Accuracy: {accuracy_rf  * 100:.2f}%")
```

```python
#applying best algo
```

```python
final = RandomForestClassifier()
final.fit(X,Y)
final_predict = final.predict(test)
```
