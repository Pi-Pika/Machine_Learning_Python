```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
iris.target_names
```

```python
df[df.target==1].head()
```

```python
df['flower_names'] = df.target.apply(lambda x : iris.target_names[x])
df.head()
test = df.drop(['flower_nammes'], axis = 'columns')
```

```python
test.head()
```
