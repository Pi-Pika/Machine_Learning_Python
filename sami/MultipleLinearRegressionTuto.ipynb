```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
df
```

```python
#data preprocessing
```

```python
median_bedrooms = df.bedrooms.median()#finding the median of bedrooms
median_bedrooms
```

```python
df.bedrooms.fillna(median_bedrooms)#filling the nan variable
```

```python
df.bedrooms = df.bedrooms.fillna(median_bedrooms)#filling the nan variable in main csv file
df
```

```python
x = df[["area"]].values #data extracting
```

```python
y = df[["bedrooms"]].values #data extracting
y
```

```python
z = df[["age"]].values #data extracting
```

```python
q = df[["price"]].values #data extracting
q
```

```python
reg = linear_model.LinearRegression() #creating an object of the algorithm
reg
```

```python
X = np.column_stack((x,y,z)) #stacking all values in a single variable
Y = q
```

```python
reg.fit(X,Y) #training the dataset
Y
```

```python
o = np.array([[5000,7,9]]) #predicting the values
reg.predict(o)
```

```python
reg.score(X,Y) #accuracy of the model: our model is 95% accurate
```
