```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices2.csv")
df
```

```python
#data preprocessing
```

```python
df[["area"]]#data extracting
```

```python
df[["price"]]#data extracting
```

```python
x = df[["area"]].values#csv to array
x
```

```python
y = df[["price"]].values#csv to array
y
```

```python
reg = linear_model.LinearRegression()#creating an object of the algorithm
reg
```

```python
reg.fit(x, y)#training the data
```

```python
reg.predict([[5000]])#predicting the y for the value of x = 5000
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
reg.score(x,y)#accuracy of the model: our model is 95% accurate
```
