```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df.head()
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('price(&)')
plt.scatter(df.year,df.income, color = "red", marker = '+')
```

```python
year = df[["year"]].values
```

```python
income = df[["income"]].values
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(year, income)
```

```python
inc2025 = float(reg.predict([[2025]]))
inc2025
```

```python
coef = float(reg.coef_)
coef
```

```python
intercept = float(reg.intercept_)
intercept
```

```python
reg.score(year, income)
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('price(&)')
plt.scatter(df.year,df.income, color = "red", marker = '+')
plt.plot(df.year,reg.predict(df[['year']]), color = "blue")
```

```python
df2 = pd.read_csv("year_canada.csv")
```

```python
p = reg.predict(df2)
p
```

```python
df2['income'] = p
df2.income
```
