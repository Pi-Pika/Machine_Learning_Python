```python
import numpy as np
import pandas as pd
import math
```

```python
df = pd.read_csv("test_scores.csv")
df
```

```python
x = df['math'].values
x
```

```python
y = df['cs'].values
y
```

```python
def gradient_descent(x,y):
    m_curr = b_curr = 0
    iterations = 0
    n = len(x)
    learning_rate = 0.0000004
    cost1 = 500
    while True:
        y_predicted = m_curr * x + b_curr
        cost = (1/n) * sum([val**2 for val in (y-y_predicted)])
        md = -(2/n)*sum(x*(y-y_predicted))
        bd = -(2/n)*sum(y-y_predicted)
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        print ("m {}, b {}, cost {} ".format(m_curr,b_curr,cost))
        iterations += 1
        if math.isclose(cost1, cost, abs_tol = 1e-20) == True:
            print(m_curr, b_curr, iterations)
            break
        cost1 = cost
```

```python
gradient_descent(x,y)
```
